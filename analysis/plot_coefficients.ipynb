```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricare i dati dal file CSV
test_dim = 2
test_num = 2
df = pd.read_csv(f"../tests/{test_dim}D/test_0{test_num}/build/forces_vs_time.csv")

# Calcolare la media dei coefficienti per ogni valore di Reynolds
re_value = 20

offset_init = 10
offset_end = -1
name_graph = "Cd" # or Cd


df["time"] = df["time"][offset_init:offset_end]
df[name_graph] = df[name_graph][offset_init:offset_end]


# Creare il grafico
plt.figure(figsize=(10, 5))
#plt.plot(df["time"], df["Cd"], label="Drag Coefficient (Cd)", linestyle="-", linewidth=2)
plt.plot(df["time"], df[name_graph], label="Lift Coefficient (" + name_graph + ")", linestyle="-", linewidth=2)

# Etichette e titolo
plt.xlabel("Time [s]")
plt.ylabel("Coefficient Value")
plt.title(f"{name_graph} coefficient over Time test case {test_dim}D-{test_num}")
plt.legend()
plt.grid(True)

# Salvare il grafico come PNG
plt.savefig(f"{name_graph}_coefficient_over_Time_{test_dim}D_Re_{re_value}_{test_num}.png", dpi=300)
plt.show()

print(df[name_graph].mean())
```
